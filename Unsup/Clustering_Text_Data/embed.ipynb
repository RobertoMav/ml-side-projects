```python
with open("data.txt", "r") as f:
    data = f.readlines()

print(data)
data = [item.strip() for item in data]
text_data = [item.split('. ')[1] for item in data]
print(text_data)
```

```python
response = co.embed(
  texts=text_data,
  model='embed-english-v3.0',
  input_type='clustering'
)

len(response.embeddings[0])
embeddings = response.embeddings
print(embeddings[0])
```

```python
dataset = []
for string_list in embeddings:
    float_list = [float(x) for x in string_list]
    numpy_array = np.array(float_list).reshape(1024)
    dataset.append(numpy_array)

# Convert the list of numpy arrays to a single numpy array if needed
dataset = np.array(dataset)
print(dataset.shape)
np.save('dataset.npy', dataset)
```
